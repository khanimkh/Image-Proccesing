3) How does the sampling rate affect the shape of the signal?

By using the low sample rate, the fast changes of the signal can not be followed. By increasing the sampling rate, the sampled signal is closer to the source signal.

4) Among the proposed sampling rates, which ones satisfy the Nyquist-Shannon theorem? In practise, what trade-off must we do when chosing a sampling rate ?

A sampling rate 𝐹𝑒 of 180 Hz et 330Hz can satisfy Nyquist-Shannon theorem. Based on the Nyquist-Shannon theorem, the sampling rate should be bigger (or equally) than 2 times of maximum frequency of signal. In this sample, the maximum frequency of signal is 82.5 then, the sampling rate 𝐹𝑒 should be bigger (or equally) thab $2 \times 82.5=165$.
In practice, the sampling switch speed and the memory budget limit the sampling frequency upper bound. The trade-off would be between the mentioned factors and the high sampling rate.
